```python
import rasterio
```

```python
import pandas as pd

file = "Tomato.tif"

with rasterio.open(file) as src:
    band = src.read(1)
    transform = src.transform
    rows, cols = band.shape

coords = []
values = []

for r in range(rows):
    for c in range(cols):
        x, y = transform * (c, r)   # pixel → coordinate
        coords.append((x, y, band[r, c]))

df = pd.DataFrame(coords, columns=["longitude", "latitude", "value"])
df.to_csv("Tomato.csv", index=False)

df.head()
```
